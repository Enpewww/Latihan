```python
# Loop, Condition, and Function 

def hitung_suhu(a):
    sum = 0
    for i in a:
        sum = sum + i
    rrs = round(sum/2,2)
    if rrs < 18:
        suhu = "dingin"
    elif rrs >= 18 and rrs <=25 :
        suhu = "sejuk"
    elif rrs > 25:
        suhu = 'panas'

    return rrs,suhu
hitung_suhu([1,2,3,4,5,6,15])
```

```python
# Contoh penggunaan function untuk tuple

def statistik_suhu(*suhu):
    list_suhu = []
    dict_suhu = {}
    for i in suhu:
        if type(i) == int or type(i) == float:
            list_suhu.append(i)
        else:
            print("Diabaikan")
    if list_suhu:
        dict_suhu['min'] = min(list_suhu)
        dict_suhu['max'] = max(list_suhu)
        dict_suhu['rata-rata'] = sum(list_suhu)/len(list_suhu)

    return dict_suhu 

statistik_suhu(20, 22.5, 'dingin', 19, 18, 30, None, 27)
print(statistik_suhu(20, 22.5, 'dingin', 19, 18, 30, None, 27))
```

```python
# Contoh penggunaan function untuk tuple, dan list

def gabungan_statistik_suhu(*suhu, **kota_suhu):
    list_suhu = []
    gabungan = {}
    pisahan ={}
    
    # Loop kondisi pada *suhu
    for i in suhu:
        if type(i) == int or type(i) == float:
            list_suhu.append(i)
        else:
            print("Diabaikan")
    if list_suhu:
        gabungan['min'] = min(list_suhu)
        gabungan['max'] = max(list_suhu)
        gabungan['rata-rata'] = round(sum(list_suhu)/len(list_suhu),2)
    else:
        print("Invalid")


    # Loop kondisi pada **kota_suhu
    for kota, suhu_list in kota_suhu.items():
        valid_suhu = []
        for nilai in suhu_list:
            if type(nilai) == int or type(nilai) == float:
                valid_suhu.append(nilai)
            else:
                print('Diabaikan')
    
        if len(valid_suhu) >= 3:                                   # pisahan harus dilakukan lewat dict (which means **kwargs) gabisa sendiri kaya gabungan
            pisahan[kota] = {"min" : min(valid_suhu), 
            'max': max(valid_suhu), 
            'rata-rata':round(sum(valid_suhu)/len(valid_suhu),2)
            }
        else:
            print("Invalid")
        
    return {'gabungan': gabungan, 'pisahan': pisahan}
    


gabungan_statistik_suhu(
    22, 25, 'salah', 30,
    Jakarta=[29, 30, 31],
    Bandung=[21, 22, 20],
    Bogor=[25, 'dingin']  # tidak valid, kurang dari 3 data numerik
)
```

```python
# 1. Sum of Even Numbers in a List
def sum_even(numbers):
    sum = 0
    for number in numbers:
        if number % 2 == 0:
            sum+=number
        else:
            print("not even")
    return sum
 
sum_even([1,2,88,44,35,23,22,68,70])
```

```python
# Soal NGC 2

# Nomor 1
def number_of_word(sentence):
    sentence = sentence.split()
    print(len(sentence))
    return 

number_of_word(sentence = "Hello World!")
number_of_word(sentence = 'Once Upon a Time in a faraway land')

# Nomor 2
def find_multiples(start_range, end_range, multiple):
    # create empty list = comprehension
    list = []
    # loop
    while start_range <= end_range:
        if start_range % multiple == 0:
            list.append(start_range)
        start_range += 1
    return list

    # calling output
print(find_multiples(start_range= 1, end_range= 20, multiple= 5))
print(find_multiples(start_range= 1, end_range= 50, multiple= 10))

# Nomor 3
def grade_converter(convert_to, **kwargs):
    
    result = {}
   
    def letter_grade(score):
        if score in range(85,101):
            return "A"
        elif score in range(70,85):
            return "B"
        elif score in range(60,70):
            return "C"
        elif score in range(50,60):
            return "D"
        elif score in range(40,50):
            return "E"
        else:
            return "F"
        
    def gpa(score):
        if score in range(85,101):
            return 4
        elif score in range(70,85):
            return 3
        elif score in range(60,70):
            return 2.5
        elif score in range(50,60):
            return 2
        elif score in range(40,50):
            return 1.5
        else:
            return 1


    for student_name, score in kwargs.items():
        if convert_to == "letter_grade":
            result  [student_name] = letter_grade(score)
        elif convert_to == "gpa":
            result  [student_name] = gpa(score)
        else:
            "Converter Invalid"

    return result
grade_converter(convert_to="gpa", Adam=62, Faiz=91)
grade_converter(convert_to='letter_grade', Albert=85, Dwi=82, Syahdan=58, Veronica=84) 

# Nomor 4
def sequence(idx):
   n = idx
   num1 = 0
   num2 = 1
   next_num = num2  
   count = 2
   result = [num1, num2]
  
   while count <= n:
      count += 1
      num1, num2 = num2, next_num
      result.append(next_num)
  
      next_num = num1 + num2
  
   print(result[idx])
sequence(0)
sequence(4)
sequence(10)
```

```python
# Kisi-kisi

# Nomor 1
def calories_burned(duration:int, exercise:str):
    if exercise == 'berlari':
      total = duration * 10
    elif exercise == 'bersepeda':
      total = duration * 8
    elif exercise == 'berenang':
      total = duration * 12
    else:
      print('inputnya salah bos!')

    print(f'kalori yang terbakar adalah {total}')

calories_burned(duration=60, exercise='berenang')

# Nomor 2
def total_session_burned_cal(*olahraga:str, each_session_duration:int):
  activity = 0 # activity set nilai 0
  for ol in olahraga: # untuk setiap string pada list olahraga, disimpan dalam ol
    if ol == 'berlari': # jika ol sama dengan string
      activity += each_session_duration * 10 # activity akan menambah each_session * 10 / activity = activity + each_session_duration * 10
    elif ol == 'bersepeda':
      activity += each_session_duration * 8
    elif ol == 'berenang':
      activity += each_session_duration * 12
    else:
      break # break itu udahan, dia bakal balik ke for lagi untuk cari ol sampai seluruh ol habis.
  print(f"olahraga anda membakar kalori sebanyak {activity}")

total_session_burned_cal('berenang', each_session_duration=10)
```
