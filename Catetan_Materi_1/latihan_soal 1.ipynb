```python
def hitung_suhu(a):
    sum = 0
    for i in a:
        sum = sum + i
    rrs = round(sum/2,2)
    if rrs < 18:
        suhu = "dingin"
    elif rrs >= 18 and rrs <=25 :
        suhu = "sejuk"
    elif rrs > 25:
        suhu = 'panas'

    return rrs,suhu
hitung_suhu([1,2,3,4,5,6,15])
```

```python
# Exercise 1

def kategori_suhu_kota(**kota):
    for value in list(kota):
        if len(kota[value]) == 7:
            sum = 0
            for i in kota[value]:
                sum = sum + i
                avg = round(sum/7,2)
                if avg <= 18:
                    suhu = "Dingin"
                elif avg > 18 and avg < 25:
                    suhu = "Sejuk"
                elif avg >= 25:
                    suhu = "Panas"
                else:
                    "not defined temperature"
                kota[value] = avg, suhu
        else:
            del kota[value]
    return kota
        
kategori_suhu_kota(Jakarta=[28, 29, 30, 27, 26, 25, 24],
Bandung=[20, 19, 18, 21, 22, 20, 21],
Bogor=[22, 23, 20]) 
```

```python
# Exercise 1 cara lain

def kategori_suhu_kota(**kota):
    result = {}
    for nama_kota, temperature in kota.items():
        if len(temperature) == 7:
            avg = round(sum(temperature)/7,2)
            if avg <= 18:
                suhu = "Dingin"
            elif avg > 18 and avg < 25:
                suhu = "Sejuk"
            elif avg >= 25:
                suhu = "Panas"
            else:
                "not defined"
            result[nama_kota] = avg, suhu
    
    return result


kategori_suhu_kota(Jakarta=[28, 29, 30, 27, 26, 25, 24],
Bandung=[20, 19, 18, 21, 22, 20, 21],
Bogor=[22, 23, 20])
            
```
