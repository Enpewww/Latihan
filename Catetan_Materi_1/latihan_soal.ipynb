```python
def statistik_suhu(*suhu):
    list_suhu = []
    dict_suhu = {}
    for i in suhu:
        if type(i) == int or type(i) == float:
            list_suhu.append(i)
        else:
            print("Diabaikan")
    if list_suhu:
        dict_suhu['min'] = min(list_suhu)
        dict_suhu['max'] = max(list_suhu)
        dict_suhu['rata-rata'] = sum(list_suhu)/len(list_suhu)

    return dict_suhu 

statistik_suhu(20, 22.5, 'dingin', 19, 18, 30, None, 27)
print(statistik_suhu(20, 22.5, 'dingin', 19, 18, 30, None, 27))
```

```python
def gabungan_statistik_suhu(*suhu, **kota_suhu):
    list_suhu = []
    gabungan = {}
    pisahan ={}
    for i in suhu:
        if type(i) == int or type(i) == float:
            list_suhu.append(i)
        else:
            print("Diabaikan")
    if list_suhu:
        gabungan['min'] = min(list_suhu)
        gabungan['max'] = max(list_suhu)
        gabungan['rata-rata'] = round(sum(list_suhu)/len(list_suhu),2)
    else:
        print("Invalid")

    for kota, suhu_list in kota_suhu.items():
        valid_suhu = []
        for nilai in suhu_list:
            if type(nilai) == int or type(nilai) == float:
                valid_suhu.append(nilai)
            else:
                print('Diabaikan')
    
        if len(valid_suhu) >= 3:                                   # pisahan harus dilakukan lewat dict (which means **kwargs) gabisa sendiri kaya gabungan
            pisahan[kota] = {"min" : min(valid_suhu), 
            'max': max(valid_suhu), 
            'rata-rata':round(sum(valid_suhu)/len(valid_suhu),2)
            }
        else:
            print("Invalid")
        
    return {'gabungan': gabungan, 'pisahan': pisahan}
    


gabungan_statistik_suhu(
    22, 25, 'salah', 30,
    Jakarta=[29, 30, 31],
    Bandung=[21, 22, 20],
    Bogor=[25, 'dingin']  # tidak valid, kurang dari 3 data numerik
)
```
