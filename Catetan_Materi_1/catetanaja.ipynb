```python
# variable-length: 
   # a. * -> tuple
   # b. ** -> dictionary

def bar(*args):
    print(args)

bar("a", "b", "c")

def bar2(*args, **kwargs):
    print(args)
    print(kwargs)

bar2(1,2,3, kw1=123, kw2="BCD")
```

```python
# Sentence Generator (concate)

def words(*args):
    print(args)
    total = ""

    for i in args:
        total = total + " " + i
    
    print(total)

words("When", "I", "was", "a", "young", "boy")
    
```

```python
# Contoh NGC 3
def grade_converter(convert_to, **kwargs):
    def letter_grade(score):
        if score in range(85,101):
            return "A"
        elif score in range(70,85):
            return "B"
        elif score in range(60,70):
            return "C"
        elif score in range(50,60):
            return "D"
        elif score in range(40,50):
            return "E"
        else:
            return "F"
        
    def gpa(score):
        if score in range(85,101):
            return 4
        elif score in range(70,85):
            return 3
        elif score in range(60,70):
            return 2.5
        elif score in range(50,60):
            return 2
        elif score in range(40,50):
            return 1.5
        else:
            return 1

    result = {}

    for student_name, score in kwargs.items():
        if convert_to == "letter_grade":
            result[student_name] = letter_grade(score)
        elif convert_to == "gpa":
            result[student_name] = gpa(score)
        else:
            "Converter Invalid"
    return result

grade_converter(convert_to="gpa", Adam=62, Faiz=91)
grade_converter(convert_to='letter_grade', Albert=90, Dwi=82, Syahdan=58, Veronica=84) 
```

```python
# Latihan simulasi LC versi gw

# Revision

'''
Using while True for each loop of the input to the list_information. Gunanya untuk selalu check input yang akan diisi user
'''

list_informasi = []

def input_data():
    print('============')
    print('Tolong isi detail informasi Anda')
    print('Isi dengan Nama Anda, ID Anda, dan Department Anda')
    print('=============')
    
    while True:
        nama_orang = (input("Masukkan nama anda :"))
        if nama_orang.isalpha():
            break
        else:
            print("Tolong input hanya dengan huruf")
    
    while True:   
        id_orang = (input("Masukkan ID anda: "))
        if id_orang.isdigit():
            break
        else:
            print('Tolong input hanya dengan angka')
        
    while True:   
        dep_orang = (input("Masukkan Dept anda: "))
        if dep_orang.isalpha():
            break
        else:
            print("Tolong input hanya dengan huruf")

    list_informasi.append([nama_orang, id_orang, dep_orang])
   
    print("Nama Anda: " + nama_orang)
    print('ID Anda: ' + str(id_orang))
    print('Department Anda: ' + dep_orang)

def check_informasi():
    no = 1
    for informasi in list_informasi:
        print(f'{no}. {informasi[0], informasi[1], informasi[2]}')
        no+=1

def detail_informasi():
    no = 1
    for informasi in list_informasi:
        print ('No | Nama | ID | Department')
        print (f'{no} | {informasi[0]} | {informasi[1]} | {informasi[2]}')
        no+=1

def menu_utama():
    while True:
        print('')
        print('')
        print('============')
        print('Tolong pilih akses file')
        print('============')
        print(' ')
        print('Pilihan terdiri dari 3:')
        print("1. Isi detail informasi Anda")
        print('2. Check Informasi')
        print('3. Akses ke Detail informasi Anda')
        print('4. Exit')    
       
        pilih_katalog = int(input('Pilih katalog yang ingin diakses: '))

        if pilih_katalog == 1:
            input_data()
        elif pilih_katalog == 2:
            check_informasi()
        elif pilih_katalog == 3:
            detail_informasi()
        elif pilih_katalog == 4:
            print("Anda telah keluar dari menu")
            break 
        else:
            print('Invalid')

menu_utama()
```
