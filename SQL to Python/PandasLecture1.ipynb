## What Cause Missing Values

Missing values can occur in datasets for various reasons, such as data collection errors, data entry issues, or missing information in the original source. Understanding the characteristics of missing values is important to determine the appropriate handling strategy. To understand the characteristics of missing data, there are three characteristics of missing values which are MCAR, MAR, and MNAR.
   
    1. Missing Completely at Random (MCAR):
        ● MCAR  refers  to  the  situation  where  the  missingness  of  data  is  unrelated  to  both  observed  and unobserved data. In other words, the missing values are randomly distributed throughout the dataset.●Handling Approach: In this case, you can simply delete the rows or columns with missing values using the dropna() method, as it won't introduce any bias in the analysis.
        - tidak ada hubungan dengan variable lain
    
    2. Missing at Random (MAR):
        ● MAR occurs when the probability of missing values depends on the observed data, but not on the unobserved data. The missingness  is related to other variables in the dataset.
        ● Handling Approach: You can use techniques like imputation, where missing values are filled in using statistical methods or algorithms based on the observed data. Common imputation methods include mean imputation, median imputation, or regression-based imputation.
        - berhubungan dengan variable lain yang ada di dataset
    
    3. Missing N at Random (MNAR):
        ● MNAR  happens  when  the  missingness  is  related  to  the  unobserved  data  or  the  missing  values themselves. The missingness is not predictable based on the observed data.●Handling Approach: Handling MNAR is challenging as it requires careful consideration of the underlying mechanisms causing missingness. Advanced techniques like multiple imputation or maximum likelihood estimation can be used to handle MNAR.
        - sistematis sulit di identifikasi, ada hubungan dengan data yang tidak diobservasi.
